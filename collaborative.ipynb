```python
def get_similar_users(user_id, top_n=5):
    if user_id not in user_similarity_df.index:
        return f"User ID '{user_id}' not found."
    
    # Sort users by similarity
    similar_users = user_similarity_df[user_id].sort_values(ascending=False)
    
    # Exclude the user itself
    similar_users = similar_users.drop(user_id)
    
    return similar_users.head(top_n)
```

```python
def get_user_id(username):
    # Find the user_id based on the given username
    user_id = df_users[df_users['username'] == username]['user_id']
    
    # Return the first match or None if no match is found
    if not user_id.empty:
        return user_id.iloc[0]  # Return the first user_id found
    else:
        return None
```

```python
def user_based_recommend(username, top_n=10):
    user_id=get_user_id(username)
    # Find similar users
    similar_users = get_similar_users(user_id, top_n=top_n)
    if isinstance(similar_users, str):  # Handle error case
        return similar_users
    
    # Get the ratings of similar users
    similar_users_ids = similar_users.index
    similar_users_ratings = pt.loc[:, similar_users_ids]  # Ratings for similar users
    
    # Get ratings of the target user
    user_ratings = pt[user_id]
    
    # Identify items that the user has not rated (where the rating is 0)
    unrated_items = user_ratings[user_ratings == 0]
    unrated_items_titles = unrated_items.index.tolist()
    unrated_items_ratings = similar_users_ratings.loc[unrated_items_titles]
    mean=unrated_items_ratings.mean(axis=1)
    recommendations = mean.sort_values(ascending=False)
    return recommendations.head(top_n).index.tolist()
```
