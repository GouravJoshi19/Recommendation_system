```python
def convert_category_to_list(row):
    category_list = [row['name']]
    return category_list
```

```python
df_posts['category']=df_posts['category'].apply(convert_category_to_list)
```

```python
def extract_keywords(data):
    if 'keywords' in data:
        return [item['keyword'] for item in data['keywords']]
    return []
```

```python
def extract_topics(data):
    if isinstance(data, dict):
        return data.get('topics_of_video', [])
    return []
```

```python
def extract_genres(data):
 if isinstance(data, dict):
        return data.get('genre', [])
 return []
```

```python
df_posts['genre']=df_posts['post_summary'].apply(extract_genres)
df_posts['keywords']=df_posts['post_summary'].apply(extract_keywords)
df_posts['topics']=df_posts['post_summary'].apply(extract_topics)
```

```python
def extract_topics_from_varied_data(data):
    if isinstance(data, dict):
        # Check for prioritized keys and return their values
        if "primary" in data:
            return data.get("primary")
        if "main_topics" in data:
            return [data.get("main_topics")[0]]
        if "main_themes" in data:
            return [data.get("main_themes")[0]]
        if "main_topic" in data:
            return [data.get("main_topic")]
        if "main_theme" in data:
            return [data.get("main_theme")]
        if "main" in data:
            return data.get("main")
        if "themes" in data:
            return data.get("themes")[0]
        if "theme" in data:
            return [data.get("theme")]
    # Return an empty list if no relevant key is found
    return data
```

```python
df_posts['topics']=df_posts['topics'].apply(extract_topics_from_varied_data)
```

```python
def extract_emotions(data):
 if isinstance(data, dict):
        return data.get('emotions', [])
 return []
```

```python
df_posts['mood'] = df_posts['post_summary'].apply(extract_emotions)
```

```python
def Convert(data):
    # Check if data is a dictionary and then look for each specific key
    if isinstance(data, dict):
        # Return values based on the available key in the dictionary
        return (
            data.get('conveyed', []) or
            data.get('expressed_emotions', []) or
            data.get('emotional_content', []) or
            data.get('moods', []) or
            data.get('moods_conveyed', []) or
            data.get('conveyed_emotions', []) or
            data.get('conveyed_moods', []) or
            data.get('emotions_expressed', []) or
            data.get('described_emotions', []) or
            data.get('expressed',[]) or
            data.get('emotional_states',[]) or
            data.get('emotions_conveyed',[])
        )
    
    # If data is not a dictionary, just return it as is
    return data
```

```python
df_posts['tags']=df_posts['category']  + df_posts['genre'] + df_posts['keywords'] 
```

```python
def convert_tags_to_string(tag_list):
    if isinstance(tag_list, __builtins__.list):
        # Flatten the list and convert all elements to strings
        flattened_list = [str(item) for sublist in tag_list for item in (sublist if isinstance(sublist,__builtins__.list) else [sublist])]
        return " ".join(flattened_list)
    return ""
```

## FUNCTION FOR THE PERSONALIZED RECOMMENDATION USING CONTENT BASED_FILTERING

```python
def Recommend(username):
    # Fetch the posts viewed by the user
    user_views = final_df[final_df['username'] == username]
    recommended_posts = []  # Initialize an empty list to store recommendations

    # Iterate through the titles the user has viewed
    for title in user_views['title']:
        # Get the index of the title in final_df
        title_index = final_df[final_df['title'] == title].index[0]
        
        # Get similarity distances for the current title
        distances = similarity[title_index]
        
        # Find the top 5 most similar posts (excluding the current post itself)
        title_list = sorted(enumerate(distances), reverse=True, key=lambda x: x[1])[1:6]
        
        # Add the recommended posts to the list
        for idx, _ in title_list:
            recommended_post = final_df.iloc[idx]['title']
            if recommended_post not in recommended_posts:  # Avoid duplicates
                recommended_posts.append(recommended_post)
    return recommended_posts 
```

```python
def get_liked_posts(username):
    likes=final_liked_df[final_liked_df['username']==username]['title'].tolist()
    return likes
```

```python
def calculate_rmse(username,recommended_posts):
    # Combine unique posts from both lists
    true_likes=get_liked_posts(username)
    all_posts = list(set(true_likes + recommended_posts))
    
    # Create binary vectors
    true_vector = [1 if post in true_likes else 0 for post in all_posts]
    recommended_vector = [1 if post in recommended_posts else 0 for post in all_posts]
    
    # Calculate RMSE
    rmse = np.sqrt(np.mean((np.array(true_vector) - np.array(recommended_vector))**2))
    return rmse
    
rmse = calculate_rmse('mike',post_recommended)
print(f"RMSE: {rmse}")
```

```python
def calculate_mae(username, recommended_posts):
    true_likes=get_liked_posts(username)
    all_posts = list(set(true_likes + recommended_posts))
    
    # Create binary vectors
    true_vector = [1 if post in true_likes else 0 for post in all_posts]
    recommended_vector = [1 if post in recommended_posts else 0 for post in all_posts]
    
    # Calculate MAE
    mae = np.mean(np.abs(np.array(true_vector) - np.array(recommended_vector)))
    return mae
    
mae = calculate_mae('mike',post_recommended)
print(f"MAE: {mae}")
```
